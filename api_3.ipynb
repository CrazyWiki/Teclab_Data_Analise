```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



esperanza_de_vida_df = pd.read_csv('/Users/victoryzelova/Desktop/TEclab/Analisis datos/API2/ARCHIVOS/esperanza_de_vida.csv', encoding='latin-1')
df_exportaciones = pd.read_csv('/Users/victoryzelova/Desktop/TEclab/Analisis datos/API2/ARCHIVOS/exportaciones.csv', encoding='latin-1')

# Filtrar los datos para el año 2025
esperanza_vida_2025 = esperanza_de_vida_df[esperanza_de_vida_df['anio'] == 2025]
# Crear el histograma
plt.figure(figsize=(10, 6))
sns.histplot(data=esperanza_vida_2025, x='mujeres', kde=True, bins=10)
plt.title('Distribución de la Esperanza de Vida de Mujeres por Provincia (2025)')
plt.xlabel('Esperanza de Vida (años)')
plt.ylabel('Frecuencia de Provincias')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Configuración del estilo del gráfico
poblacion_df = pd.read_csv('/Users/victoryzelova/Desktop/TEclab/Analisis datos/API2/ARCHIVOS/poblacion.csv', encoding='latin-1')
hogares_df = pd.read_csv('/Users/victoryzelova/Desktop/TEclab/Analisis datos/API2/ARCHIVOS/hogares_viviendas_superficie.csv', encoding='latin-1')

# Fusionar los dos dataframes por provincia
# Se asume que los datos de hogares son del último censo disponible y se comparan con la población de un año cercano
poblacion_reciente = poblacion_df[poblacion_df['anio'] == poblacion_df['anio'].max()]
datos_combinados = pd.merge(poblacion_reciente, hogares_df, on='provincia')

# Crear el diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos_combinados, x='viviendas_particulares_habitadas', y='poblacion_total')
plt.title('Relación entre Viviendas y Población por Provincia')
plt.xlabel('Cantidad de Viviendas Particulares Habitadas')
plt.ylabel('Población Total')
plt.grid(True)
plt.show()
```

```python
empleo_df = pd.read_csv('/Users/victoryzelova/Desktop/TEclab/Analisis datos/API2/ARCHIVOS/empleo.csv', encoding='latin-1')

# Filtrar los datos para la provincia de Buenos Aires
empleo_ba = empleo_df[empleo_df['provincia'] == 'Buenos Aires'].copy()

# Convertir las columnas de año y mes a un formato de fecha
empleo_ba['fecha'] = pd.to_datetime(empleo_ba['anio'].astype(str) + '-' + empleo_ba['mes'].str.split('-').str[1], format='%Y-%y')


# Crear el gráfico de línea
plt.figure(figsize=(12, 6))
sns.lineplot(data=empleo_ba, x='fecha', y='empleados_registrados_miles')
plt.title('Evolución del Empleo Registrado en Buenos Aires (2009-2018)')
plt.xlabel('Año')
plt.ylabel('Empleados Registrados (en miles)')
plt.grid(True)
plt.show()
```
